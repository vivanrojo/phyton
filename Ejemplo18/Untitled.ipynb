```python
import os, sys

def crear_tabla():
    print('[1] CREAR TABLA TRABAJADOR')
    print('--------------------------')

    os.system('pause')

def insertar():
    print('[2] INSERTAR REGISTROS')
    print('----------------------')

    os.system('pause')

def mostrar():
    print('[3] MOSTRAR TODOS LOS REGISTROS')
    print('-------------------------------')

    os.system('pause')

def actualizar():
    print('[4] ACTUALIZAR UN REGISTRO')
    print('--------------------------')

    os.system('pause')

def eliminar():
    print('[5] ELIMINAR UN REGISTRO')
    print('------------------------')

    os.system('pause')

def salir():
    os.system('cls')
    print('GRACIAS POR SU VISITA')
    sys.exit()

def default():
    pass

menu_dic = {
    '1': crear_tabla,
    '2': insertar,
    '3': mostrar,
    '4': actualizar,
    '5': eliminar,
    '0': salir
}

def main():
    while True:
        os.system("cls")
        print('MENU')
        print('----')
        print('[1] CREAR TABLA TRABAJADOR')
        print('[2] INSERTAR REGISTROS')
        print('[3] MOSTRAR TODOS LOS REGISTROS')
        print('[4] ACTUALIZAR UN REGISTRO')
        print('[5] ELIMINAR UN REGISTRO')
        print('[0] SALIR')

        opcion = input('INGRES OPCION? ')
        funcion = menu_dic.get(opcion,default)
        os.system('cls');funcion()

if __name__ == "__main__":
   main()
 
```
